```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.preprocessing import StandardScaler

# Load IRIS dataset
iris = load_iris()
X = iris.data
y = iris.target

# Standardize features
sc = StandardScaler()
X_scaled = sc.fit_transform(X)

# Perform PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Perform LDA
lda = LinearDiscriminantAnalysis(n_components=2)
X_lda = lda.fit_transform(X, y)  # Note: LDA expects unscaled X

# Create subplots
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(13.5, 4))

# Plot PCA
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=y, palette='Set1', ax=ax[0], legend="full")
ax[0].set_title("PCA of IRIS dataset", fontsize=15, pad=15)
ax[0].set_xlabel("PC1", fontsize=12)
ax[0].set_ylabel("PC2", fontsize=12)

# Plot LDA
sns.scatterplot(x=X_lda[:, 0], y=X_lda[:, 1], hue=y, palette='Set1', ax=ax[1], legend="full")
ax[1].set_title("LDA of IRIS dataset", fontsize=15, pad=15)
ax[1].set_xlabel("LD1", fontsize=12)
ax[1].set_ylabel("LD2", fontsize=12)

# Save the figure
plt.tight_layout()
plt.savefig('PCA_vs_LDA.png', dpi=80)
plt.show()
```
