```python
from scipy import stats
import pandas as pd
import numpy as np
import pylab
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew

%matplotlib inline

# Create a skewed distribution
skewed_dist = stats.loggamma.rvs(5, size=10000) + 5

# Create a normal distribution
normal_dist = np.random.normal(0, 1, 10000)

# Calculate the skewness of the normal distribution and plot the distributions
sns.histplot(normal_dist, kde=True)
plt.title("Normal Distribution")
plt.show()

print("Skewness for the normal distribution:", skew(normal_dist))

sns.histplot(skewed_dist, kde=True)
plt.title("Skewed Distribution")
plt.show()

print("Skewness for the skewed distribution:", skew(skewed_dist))
```
