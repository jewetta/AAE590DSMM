```python
from scipy import stats
import pandas as pd
import numpy as np
import pylab
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew

%matplotlib inline

# Create a skewed distribution
skewed_dist = stats.loggamma.rvs(5, size=10000) + 5

# Create a normal distribution
normal_dist = np.random.normal(0, 1, 10000)

# Calculate the skewness of the normal distribution and plot the distributions
sns.histplot(normal_dist, kde=True)
plt.title("Normal Distribution")
plt.show()

print("Skewness for the normal distribution:", skew(normal_dist))

sns.histplot(skewed_dist, kde=True)
plt.title("Skewed Distribution")
plt.show()

print("Skewness for the skewed distribution:", skew(skewed_dist))

# Q-Q plots for normality checking
stats.probplot(normal_dist, dist="norm", plot=pylab)
pylab.title("Q-Q Plot for Normal Distribution")
pylab.show()

stats.probplot(skewed_dist, dist="norm", plot=pylab)
pylab.title("Q-Q Plot for Skewed Distribution")
pylab.show()

# Apply Box-Cox Transformation to normalize skewed data
skewed_box_cox, lambda_value = stats.boxcox(skewed_dist)

# Plot transformed data
sns.histplot(skewed_box_cox, kde=True)
plt.title("Box-Cox Transformed Skewed Distribution")
plt.show()

# Q-Q plot for Box-Cox transformed data
stats.probplot(skewed_box_cox, dist="norm", plot=pylab)
pylab.title("Q-Q Plot for Box-Cox Transformed Data")
pylab.show()

print("Lambda parameter for Box-Cox Transformation is:", lambda_value)
```
